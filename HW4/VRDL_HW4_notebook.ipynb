# Introduction 

In this homework the model is trained on the tiny pascal voc dataset with 1349 training and tested on the testing data with 100 test images. 20 common object classes exist and the biggest problem here is that the dataset is so small, which means that overfitting is an issue that needs to be addressed here. The key ideas here are to avoid overfitting with too many epochs and also to artificially increase the dataset by increasing the number of data samples with data augmentation/modification. 

**Methodology:**

After researching for possible approaches to implement, I decided to utilize the Mask R-CNN network architecture for this task. The network implementation from the linked sources is being utilized. 

**Data pre-processing:**

The datasets have been loaded into Google Colab and were extracted. Then the annotation data for the images was loaded.
Furthermore in order to cope with the overfit problem, images were augmented with data modifications.
This means image were flipped to artificially increase the available testing data.

```python
import numpy as np
from itertools import groupby

def binary_mask_to_rle(binary_mask):
    rle = {'counts': [], 'size': list(binary_mask.shape)}
    counts = rle.get('counts')
    for i, (value, elements) in enumerate(groupby(binary_mask.ravel(order='F'))):
        if i == 0 and value == 1:
            counts.append(0)
        counts.append(len(list(elements)))
    return rle
```

```python
#Given: File_name, prediction

TEST_IMAGE_DIR = 'test_images'
# Each instance consists of keys: "image_id", "score", "category_id", "segmentation" where segmentation is the encoded mask in RLE
# Process: go through all test images one by one, process each image and then extract instances for submission

import os
os.remove(TEST_IMAGE_DIR+"/"+"desktop.ini") 

# Iterate through the testing data
image_counter = 0
for filename in os.listdir(TEST_IMAGE_DIR):
    print("Image number:", image_counter)
    image_counter=image_counter+1
    print(filename)
    # load each test image
    image = cv2.imread("test_images/"+filename)
    #image.show()
    # Run detection on this image
    results = model.detect([image], verbose=1)

    # Visualize results
    r = results[0]
    #visualize.display_instances(image, r['rois'], r['masks'], r['class_ids'], class_names, r['scores'])

    # for submission store iamge_id and create list of instances, categories, scores and masks
    # Store all dictionaries with the instances in the submission_list
    submission_list = []

    # Image_id
    for i in range(len(test_data['images'])):
      if(test_data['images'][i]['file_name'] == filename):
        image_id = i

    # Categories of instanecs
    category_list = r['class_ids']

    # Scores
    score_list = r['scores']

    # Masks
    segmentation_masks = binary_mask_to_rle(r['masks'])
    
    dict_ = {}

    # For all instances detected in this image, create a dictionary and add it to the list
    for instance in range(len(score_list)):
      dict_['image_id'] = image_id
      dict_['category_id'] = category_list[instance]
      dict_['score'] = score_list[instance]
      dict_['segmentation'] = segmentation_masks
    submission_list.append(dict_)


    with open("0845058.json", 'w') as f:
      f.write(str(submission_list))
```
